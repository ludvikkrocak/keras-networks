```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from scipy.signal import convolve
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from kerastuner.tuners import RandomSearch
```

```python
# Zpracování atributu 'Depth'
train['Depth'] = (train['Depth'] == 'Topsoil').astype(int)
test['Depth'] = (test['Depth'] == 'Topsoil').astype(int)
```

```python
# Nastavení vstupních dat a cílových proměnných
y = train.iloc[:, -5:].values
X = train.iloc[:, 2:-5].values
```

```python
# Standardizace vstupních dat
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Definice Hammingova okna
window = np.hamming(16)

# Aplikace Hammingova okna na data
smoothed_data = np.apply_along_axis(lambda x: convolve(x, window, mode='same'), axis=1, arr=X)

# Vyberte konkrétní sloupce z transformovaných dat
selected_columns = np.concatenate((smoothed_data[:, :40], smoothed_data[:, 100:]), axis=1)
```

```python
# Dělení dat na tréninkovou a testovací množinu
X_train, X_test, y_train, y_test = train_test_split(selected_columns, y, random_state=42, test_size=0.2)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Definice funkce pro vytvoření modelu s early stopping
def build_model(hp):
    model = Sequential()
    model.add(Dense(units=hp.Int('units_1', min_value=64, max_value=256, step=64), activation='relu', input_dim=X_train.shape[1]))
    model.add(Dropout(rate=hp.Float('dropout_1', min_value=0.2, max_value=0.8, step=0.2)))
    model.add(Dense(units=hp.Int('units_2', min_value=32, max_value=256, step=32), activation='relu'))
    model.add(Dense(units=hp.Int('units_3', min_value=32, max_value=256, step=32), activation='relu'))
    model.add(Dropout(rate=hp.Float('dropout_2', min_value=0.2, max_value=0.8, step=0.2)))
    model.add(Dense(units=hp.Int('units_4', min_value=32, max_value=256, step=32), activation='relu'))
    model.add(Dense(5, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

    # Early stopping
    callbacks = [
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=10,
            restore_best_weights=True
        )
    ]

    return model
```

```python
# Vytvoření tuneru s použitím callbacks pro early stopping
tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=50,
    directory='my_dir',
    project_name='my_project'
)
```

```python
# Spuštění vyhledávání hyperparametrů
tuner.search(X_train, y_train, validation_split=0.2, epochs=200)
```

```python
# Získání nejlepší konfigurace a modelu
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
best_model = tuner.get_best_models(num_models=1)[0]
```

```python
best_hyperparameters = best_hps.get_config()

print("Nejlepší konfigurace hyperparametrů:")
print(best_hyperparameters)
```

```python
# Vytvoření instance EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Natrénování modelu na tréninkových datech
best_model.fit(X_train, y_train, validation_split=0.2, epochs=100, callbacks=[early_stopping])
```

```python
# Ohodnocení modelu na testovacích datech
test_loss, test_accuracy = best_model.evaluate(X_test, y_test)
print(f"Test accuracy: {test_accuracy}")
```
